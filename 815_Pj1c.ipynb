```python
#Question 1
%matplotlib inline
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x0 = 0                    #initial condition
n  = 50                 #interval values: 1, 10, 100, 1000, 10000
y0 = 0                    #initial value
a = -52.419               #coefficients
b = 58.252
c = 150.39
Vg = 20.0
w = 50
```

```python
def f(x,y):
    return (Vg*np.cos(w*x) - ((2*a*x)+(4*b*x**3)+(6*c*x**5)))

def Runge_Kutta(f, x0, xf, n, y0):
    RKx = np.zeros(n)
    RKy = np.zeros(n)

    #h = (xf - x0) / n
    h = 0.005
    x = x0
    y = y0

    RKx[0] = x
    RKy[0] = y

    for i in range(0, n):
        k1 = h * f(x, y)
        k2 = h * f(x + h / 2, y + k1 / 2)
        k3 = h * f(x + h / 2, y + k2 / 2)
        k4 = h * f(x + h, y + k3)

        x = x0 + (i + 1) * h
        y += (k1 + 2 * k2 + 2 * k3 + k4) / 6

        RKx[i] = x
        RKy[i] = y

    #return [RKx, RKy]
    plt.plot(RKx,RKy)
    plt.xlabel('t')
    plt.ylabel('Q')
    
```
