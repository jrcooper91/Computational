```python
from random import seed
from random import random
import numpy as np
import matplotlib.pyplot as plt 
```

```python
"average value of x"
seed(0)
random_walk = list()
random_walk.append(-1 if random() < 0.5 else 1)
for i in range(0, 5000,1):
    movement = -1 if random() < 0.5 else 1
    value = random_walk[i-1] + movement
    random_walk.append(value)
    avg = np.mean(random_walk)
    #print(avg)
    plt.scatter(i,avg,s=1)
plt.xlabel('time')    
plt.ylabel('x value average')
plt.show()
```

```python
ent1=[]
seed(0)
random_walk = list()
random_walk.append(-1 if random() < 0.5 else 1)
for i in range(1, 5000,1):
    movement = -1 if random() < 0.5 else 1
    value = random_walk[i-1] + movement
    #random_walk.append(value)
    p_i=(i+1)/i #density
    ent=-(1/i)*(p_i*np.log(p_i)+np.log(i)) #entropy
    random_walk.append(ent)
    entropy=np.sum(random_walk)
    plt.scatter(i,entropy)
plt.xlabel('time')  
plt.ylabel('entropy')
plt.show()
```
